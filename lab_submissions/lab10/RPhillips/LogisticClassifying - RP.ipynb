```python
def makeplot_train(classifier):
    x_min, x_max = X_train[:, 0].min() - .5, X_train[:, 0].max() + .5
    y_min, y_max = X_train[:, 1].min() - .5, X_train[:, 1].max() + .5
    
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))
    
    Z = classifier.predict(np.c_[xx.ravel(), yy.ravel()])
    
    Z = Z.reshape(xx.shape)
    pl.figure(1, figsize=(15,11))
    pl.pcolormesh(xx, yy, Z, cmap=pl.cm.Paired)
    
    # Plot the TRAINING points
    pl.scatter(X_train[:,0], X_train[:, 1], c=Y_train, edgecolors='k', s=100, cmap=pl.cm.Paired)
    pl.xlabel('Sepal length')
    pl.ylabel('Sepal width')
    
    # Plot the TESTING points
    # pl.scatter(X_test[:,0], X_test[:, 1], c=Y_test, edgecolors='k', s=100, cmap=pl.cm.Paired)
    # pl.xlabel('Sepal length')
    # pl.ylabel('Sepal width')
    
    pl.xlim(xx.min(), xx.max())
    pl.ylim(yy.min(), yy.max())
    pl.xticks(())
    pl.yticks(())
    pl.title(classifier.__module__)
    pl.size(5)
    
    pl.show()
```

```python
from sklearn.naive_bayes import GaussianNB, BernoulliNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
```

```python
# Create five new variables for the new classifiers

gnb = GaussianNB()
bnb = BernoulliNB()
lr = LogisticRegression()
dt = DecisionTreeClassifier(criterion='entropy', max_depth=3,min_samples_leaf=5)
knn = KNeighborsClassifier()

# General rule for applying classifiers: variable = variable.fit(X_train,y_train)

gnb.fit(X_train, Y_train)
bnb.fit(X_train, Y_train)
lr.fit(X_train, Y_train)
dt.fit(X_train, Y_train)
knn.fit(X_train, Y_train)

# Here I am calling the scores for each classifier

gnb_score = gnb.score(X_test, Y_test)
bnb_score = bnb.score(X_test, Y_test)
lr_score = lr.score(X_test, Y_test)
dt_score = dt.score(X_test, Y_test)
knn_score = knn.score(X_test, Y_test)

# Next step after building this is to print the score for each classifier
 
print "Gaussian: %0.2f%%" %(gnb_score * 100.0)
print "Bernulli: %0.2f%%" %(bnb_score * 100.0)
print "Linear Regression: %0.2f%%" %(lr_score * 100.0)
print "Decision Tree: %0.2f%%" %(dt_score * 100.0)
print "KNN: %0.2f%%" %(knn_score * 100.0)

```

```python
gnb = GaussianNB()
bnb = BernoulliNB()
lr = LogisticRegression()
dt = DecisionTreeClassifier(criterion='entropy', max_depth=3,min_samples_leaf=5)
knn = KNeighborsClassifier()

# Here I am creating an empty list and defining a variable to allow for multiple percentage splits in the test_size
p = [.05, .1, .25, .3, .5, .75]
gnb_score_list = []
bnb_score_list = []
lr_score_list = []
dt_score_list = []
knn_score_list = []

# Here I created a for loop function that will list the different percentages for the test_size split
for i in p:
    X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=i)
    gnb.fit(X_train, Y_train)
    bnb.fit(X_train, Y_train)
    lr.fit(X_train, Y_train)
    dt.fit(X_train, Y_train)
    knn.fit(X_train, Y_train)

# Call Scores for each one

    gnb_score = gnb.score(X_test, Y_test)
    gnb_score_list.append(gnb_score)
    
    bnb_score = bnb.score(X_test, Y_test)
    bnb_score_list.append(bnb_score)
    
    lr_score = lr.score(X_test, Y_test)
    lr_score_list.append(lr_score)
    
    dt_score = dt.score(X_test, Y_test)
    dt_score_list.append(dt_score)
    
    knn_score = knn.score(X_test, Y_test)
    knn_score_list.append(knn_score)

# Print all the scores at each respective test_size split and separate them by dashes for visual ease
 
    print "Precentage Test Size: %0.2f%%" %(i * 100.0)
    print "Gaussian: %0.2f%%" %(gnb_score * 100.0)
    print "Bernulli: %0.2f%%" %(bnb_score * 100.0)
    print "Linear Regression: %0.2f%%" %(lr_score * 100.0)
    print "Decision Tree: %0.2f%%" %(dt_score * 100.0)
    print "KNN: %0.2f%%" %(knn_score * 100.0)
    print "-" *28
```

```python
# Create a for loop to run through each classification test at each percentage
# Print out the best score for each classification and at which test_split that percentage was achieved
# This is accomplished by starting the max_index and max_score for each at zero and running through all the tests

bnb_max_index = 0
gnb_max_index = 0
lr_max_index = 0
dt_max_index = 0
knn_max_index = 0

bnb_max_score = 0
gnb_max_score = 0
lr_max_score = 0
dt_max_score = 0
knn_max_score = 0

for i in range(0,len(p)):
    if bnb_score_list[i] > bnb_max_score:
        bnb_max_score = bnb_score_list[i]
        bnb_max_index = i
    if gnb_score_list[i] > gnb_max_score:
        gnb_max_score = gnb_score_list[i]
        gnb_max_index = i
    if lr_score_list[i] > lr_max_score:
        lr_max_score = lr_score_list[i]
        lr_max_index = i
    if dt_score_list[i] > dt_max_score:
        dt_max_score = dt_score_list[i]
        dt_max_index = i
    if knn_score_list[i] > knn_max_score:
        knn_max_score = knn_score_list[i]
        knn_max_index = i


print "Gaussian: %0.2f%% at %0.2f%%" %(gnb_max_score * 100.0, p[gnb_max_index])
print "Bernulli: %0.2f%% at %0.2f%%" %(bnb_max_score * 100.0, p[bnb_max_index])
print "Linear Regression: %0.2f%% at %0.2f%%" %(lr_max_score * 100.0, p[lr_max_index])
print "Decision Tree: %0.2f%% at %0.2f%%" %(dt_max_score * 100.0, p[dt_max_index])
print "KNN: %0.2f%% at %0.2f%%" %(knn_max_score * 100.0, p[knn_max_index])
print "-" *20
```
